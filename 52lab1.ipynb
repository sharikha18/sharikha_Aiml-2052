```python
import numpy as np
list1 = [1, 2, 3]
arr1 = np.array(list1)
print(arr1)
```

```python
list2 = [3,2,1]
arr2 = np.array(list2)
print(arr2)
```

```python
print("array 2 mul by array 1 :",arr1*arr2)
print("array 2 divided by array 1 :",arr2/arr1)
print("array 2 raised to power of array 1 :",arr2**arr1)
print(f"{arr2} raised to the power of {arr1} :,{arr2**arr1}")
```

```python
#sine function
print("sine:",np.sin(arr1))
#log
print("natural log:",np.log(arr1))
print("base-10 log:",np.log10(arr1))
print("bas-2 log:",np.log2(arr1))
#exponential
print("exponential:", np.exp(arr1))
```

part-2 visualize the datasets or data using matplotlib

```python
ppl = ['ann','brandon','chen','david', 'emily','farook','gagan','hamish','imran','julio','katherune','lily']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
wgt = [55,35,77,68,70,60,72,69,18,65,82,48]
hgt = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(wgt,hgt)
plt.title("rltnship bwetween wgt and hgt of patients")
plt.ylabel("hgt in meters")
plt.xlabel("wgt of patients in kgs")
plt.show()
```

```python
plt.scatter(age,hgt)
plt.title("rltnship bwetween age and hgt of patients")
plt.ylabel("hgt in meters")
plt.xlabel("age of patients")
plt.show()
```

```python
plt.scatter(age,wgt)
plt.title("rltnship bwetween age and wgt of patients")
plt.ylabel("wgt of patients in kgs")
plt.xlabel("age of patients")
plt.show()
```

```python
#bar plot
plt.figure(figsize=(12,4))
plt.title("peoples wgt in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=ppl,height=wgt,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("ppl",fontsize=15)

plt.ylabel("wgt(in kgs)",fontsize=15)
plt.show()
```
